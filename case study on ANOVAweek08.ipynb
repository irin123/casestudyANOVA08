# ● Check whether all the zones generate the same amount of sales.


# Anova test

```python
fvalue,pvalue = stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
print(fvalue,pvalue)
alpha = 0.05
if pvalue > alpha:
   print("Accept the null hypothesis")
elif pvalue < alpha:
    print("Reject the null hypothesis")
```
